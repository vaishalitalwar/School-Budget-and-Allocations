```python
import pandas as pd
from re import sub
from decimal import Decimal
```

```python
def moneyToFloat(money):
    value = Decimal(sub(r'[^\d.]', '', money))
    return value
```

```python
budget = pd.read_csv("../data/budget_all.csv")
```

```python
# Converting to float
budget['amountNum'] = budget['amount'].apply((lambda x: moneyToFloat(x)))
budget.head()
```

```python
#Read all enrollment data
import glob
import os

path = '../data/enrollment/' 
all_files = glob.glob("../data/enrollment/*.xlsx")
print(all_files)
all_enrol = []

for filename in all_files:
    print(filename)
    if '$' not in filename:
        df = pd.read_excel(filename)
        all_enrol.append(df)

enrollments = pd.concat(all_enrol, axis=0, ignore_index=True)
enrollments
```

```python
enrollments = enrollments[["SCHOOL YEAR","STATE LOCATION ID","PK12 TOTAL"]]
enrollments
```

```python
enrollments["year"] = enrollments["SCHOOL YEAR"].str.split('-').str[0].astype(int)
enrollments
```

```python
dbn_mapping = pd.read_excel("../data/enrollment/$20192020StateAccountabilityStatus_ForDBNJoin.xlsx")

dbn_mapping = dbn_mapping[["BEDS","DBN"]]
dbn_mapping["School Code"] = dbn_mapping["DBN"].str[2:]
dbn_mapping
```

```python
# enrollment dataframe
dbn_enrol = dbn_mapping.merge(enrollments,left_on="BEDS",right_on="STATE LOCATION ID",how="inner")
dbn_enrol = dbn_enrol[["School Code", "year", "STATE LOCATION ID", "PK12 TOTAL"]]
dbn_enrol.head()
```

```python
# Defining teaching budget categories
# For now, defining leadership, teachers, paraprofessionals only - we can come up with other categories.
teaching_budget = budget[budget['budget_category'].isin(['Leadership', 'Paraprofessionals', 'Classroom Teacher', 'Homeroom Teacher', 'Elementary Cluster/Quota', 'Cluster/Quota Teacher'])]
teaching_budget['budget_category'] = teaching_budget['budget_category'].replace(['Classroom Teacher', 'Homeroom Teacher', 'Elementary Cluster/Quota', 'Cluster/Quota Teacher'], 'Teacher')
teaching_budget.sample(10)
```

```python
# Summing positions
positions_per_school = teaching_budget.groupby(by = ['location_code', 'fiscal_year', 'budget_category'])['num_positions'].sum().reset_index()
positions_per_school
```

```python
positions_per_school["prev year"]=positions_per_school["fiscal_year"]-1
```

```python
dbn_enrol["year"] = dbn_enrol["year"].astype(int)
```

```python
budget_pos_enroll = positions_per_school.merge(dbn_enrol,left_on=("location_code","fiscal_year"),right_on=("School Code","year"),how="left")
budget_pos_enroll = budget_pos_enroll.drop(columns=["year","School Code"])
budget_pos_enroll = budget_pos_enroll.rename(columns={"PK12 TOTAL":"Total Enrollment"})
budget_pos_enroll.loc[0:20, :]
```

```python
budget_pos_enroll = budget_pos_enroll.merge(dbn_enrol,left_on=("location_code","prev year"),right_on=("School Code","year"),how="left")
budget_pos_enroll = budget_pos_enroll.drop(columns=["year","School Code","STATE LOCATION ID_x","STATE LOCATION ID_y"])
budget_pos_enroll = budget_pos_enroll.rename(columns={"PK12 TOTAL":"Previous Year Total Enrollment"})
budget_pos_enroll
```

```python
columns_titles = ["location_code","fiscal_year", "budget_category", "num_positions", "Total Enrollment", "prev year", "Previous Year Total Enrollment"]
budget_pos_enroll=budget_pos_enroll.reindex(columns=columns_titles)
budget_pos_enroll
```

```python
budget_pos_enroll.to_csv("../data/staff numbers and enrollment 2017-2023.csv")
```
