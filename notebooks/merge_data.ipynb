```python
import pandas as pd
from re import sub
from decimal import Decimal
```

```python
def moneyToFloat(money):
    value = Decimal(sub(r'[^\d.]', '', money))
    return value
```

```python
budget = pd.read_csv("../data/budget_all.csv")
```

```python
# Converting to float
budget['amountNum'] = budget['amount'].apply((lambda x: moneyToFloat(x)))
budget.head()
```

## Reading + Processing Enrollment Data

```python
#Read all enrollment data
import glob
import os

path = '../data/enrollment/' 
all_files = glob.glob("../data/enrollment/*.xlsx")
print(all_files)
all_enrol = []

for filename in all_files:
    print(filename)
    if '$' not in filename:
        df = pd.read_excel(filename)
        all_enrol.append(df)

enrollments = pd.concat(all_enrol, axis=0, ignore_index=True)
enrollments
```

```python
enrollments = enrollments[["SCHOOL YEAR","STATE LOCATION ID","PK12 TOTAL"]]
enrollments
```

```python
enrollments["year"] = enrollments["SCHOOL YEAR"].str.split('-').str[0].astype(int)
enrollments
```

```python
dbn_mapping = pd.read_excel("../data/enrollment/$20192020StateAccountabilityStatus_ForDBNJoin.xlsx")

dbn_mapping = dbn_mapping[["BEDS","DBN"]]
dbn_mapping["School Code"] = dbn_mapping["DBN"].str[2:]
dbn_mapping
```

```python
# enrollment dataframe
dbn_enrol = dbn_mapping.merge(enrollments,left_on="BEDS",right_on="STATE LOCATION ID",how="inner")
dbn_enrol = dbn_enrol[["School Code", "year", "STATE LOCATION ID", "PK12 TOTAL"]]
dbn_enrol.head()
```

## Staff Categories and Student Enrollment

```python
# Defining staff categories
# For now, defining leadership, teachers, paraprofessionals only - we can come up with other categories.
teaching_budget = budget[budget['budget_category'].isin(['Leadership', 'Paraprofessionals', 'Classroom Teacher', 'Homeroom Teacher', 'Elementary Cluster/Quota', 'Cluster/Quota Teacher'])]
teaching_budget['budget_category'] = teaching_budget['budget_category'].replace(['Classroom Teacher', 'Homeroom Teacher', 'Elementary Cluster/Quota', 'Cluster/Quota Teacher'], 'Teacher')
teaching_budget.sample(10)
```

```python
# Summing positions
positions_per_school = teaching_budget.groupby(by = ['location_code', 'fiscal_year', 'budget_category'])['num_positions'].sum().reset_index()
positions_per_school
```

```python
positions_per_school["prev year"]=positions_per_school["fiscal_year"]-1
```

```python
dbn_enrol["year"] = dbn_enrol["year"].astype(int)
```

```python
budget_pos_enroll = positions_per_school.merge(dbn_enrol,left_on=("location_code","fiscal_year"),right_on=("School Code","year"),how="left")
budget_pos_enroll = budget_pos_enroll.drop(columns=["year","School Code"])
budget_pos_enroll = budget_pos_enroll.rename(columns={"PK12 TOTAL":"Total Enrollment"})
budget_pos_enroll.loc[0:20, :]
```

```python
budget_pos_enroll = budget_pos_enroll.merge(dbn_enrol,left_on=("location_code","prev year"),right_on=("School Code","year"),how="left")
budget_pos_enroll = budget_pos_enroll.drop(columns=["year","School Code","STATE LOCATION ID_x","STATE LOCATION ID_y"])
budget_pos_enroll = budget_pos_enroll.rename(columns={"PK12 TOTAL":"Previous Year Total Enrollment"})
budget_pos_enroll
```

```python
columns_titles = ["location_code","fiscal_year", "budget_category", "num_positions", "Total Enrollment", "prev year", "Previous Year Total Enrollment"]
budget_pos_enroll=budget_pos_enroll.reindex(columns=columns_titles)
budget_pos_enroll
```

```python
budget_pos_enroll.to_csv("../data/Task Cards/staff numbers and enrollment 2017-2023.csv")
```

## Allocation Categories Ngram

```python
# Unigram on other and none
all_merg = pd.read_csv("../data/Task Cards/allocation_category_2017_2023.csv")
all_merg

unigram = {}
for i in all_merg.loc[(all_merg["allocation_category_group"] == "Other") |  all_merg["allocation_category_group"].isnull(), "allocation_category"].unique():
    temp = i.replace('-','').split()
    for j in i.split():
        if j in unigram:
            unigram[j]+=1
        else:
            unigram[j]=1



from collections import Counter
unigram_freq = Counter(unigram)
unigram_freq
```

```python
# Bigram on everything
all_merg = pd.read_csv("../data/Task Cards/allocation_category_2017_2023.csv")
all_merg

bi_gram = {}
for i in all_merg["allocation_category"].unique():
    temp = i.replace('-',"").split()
    if len(temp) <2:
        continue
    else:
        for j in range(len(temp)-1):
            temp_words = " ".join([temp[j],temp[j+1]])
            if temp_words in bi_gram:
                bi_gram[temp_words]+=1
            else:
                bi_gram[temp_words]=1


bi_gram_freq = Counter(bi_gram)
bi_gram_freq
```

```python
# bi-gram on none and other
bi_gram = {}
for i in other:# all_merg.loc[(all_merg["allocation_category_group"] == "Other") |  all_merg["allocation_category_group"].isnull(), "allocation_category"].unique():
    temp = i.replace('-','').split()
    if len(temp) <2:
        continue
    else:
        for j in range(len(temp)-1):
            temp_words = " ".join([temp[j],temp[j+1]])
            if temp_words in bi_gram:
                bi_gram[temp_words]+=1
            else:
                bi_gram[temp_words]=1

bi_gram_freq = Counter(bi_gram)
bi_gram_freq
```

```python
# tri-gram
tri_gram = {}
for i in all_merg.loc[(all_merg["allocation_category_group"] == "Other") |  all_merg["allocation_category_group"].isnull(), "allocation_category"].unique():
    temp = i.replace('-','').split()
    if len(temp) <3:
        continue
    else:
        for j in range(len(temp)-2):
            temp_words = " ".join([temp[j],temp[j+1],temp[j+2]])
            if temp_words in bi_gram:
                tri_gram[temp_words]+=1
            else:
                tri_gram[temp_words]=1

tri_gram_freq = Counter(tri_gram)
tri_gram_freq
```

```python
data = all_merg.loc[(all_merg["allocation_category_group"] == "Other") |  all_merg["allocation_category_group"].isnull(), "allocation_category"].unique()

fsf ={}
c4e={}
title_1={}
rs={}
pres={}
sbss={}
other={}
arpa={}
pubscl={}
pg={}
nv = {}

def add_to_dict(d,s):
    if s in d:
        d[s] += 1
    else:
        d[s] =1
    return d


def compare(l,s):
    for i in l:
        if i.lower() in s:
            return True
    return False


for i in data:
    temp = i.lower()
    if compare(['fair student funding', "fsf"],temp):
        fsf = add_to_dict(fsf,i)
    elif compare(['Contract for Excellence', "c4e"], temp):
        c4e = add_to_dict(c4e,i)
    elif compare( ['Title I ', "Title 1 ", "TI "], temp):
        title_1 = add_to_dict(title_1,i)
    elif compare( ['rs', "Related Services"], temp):
        rs = add_to_dict(rs,i)
    elif compare( ['Pre-K', '3-K', '4-K'], temp):
        pres = add_to_dict(pres,i)
    elif compare( ['Sbst', "School Support"], temp):
        sbss = add_to_dict(sbss,i)
    elif compare( ['ARPA'], temp):
        arpa = add_to_dict(arpa,i)
    elif compare( ['Pub Schl '], temp):
        pubscl = add_to_dict(pubscl,i)
    elif compare( ['Private Grant ', 'Private PTA Grant '], temp):
        pg = add_to_dict(pg,i)
    elif compare(['New Visions ', 'NEW VISIONS'], temp):
        nv = add_to_dict(nv, i)
    else:
        other[i]=1
    
        
# Other potential categories: Private Grant, (see if you need to put PTA in a separate place), State Misc    
```
